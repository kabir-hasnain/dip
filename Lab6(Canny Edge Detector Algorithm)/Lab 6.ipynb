```python
import cv2
import numpy as np
from IPython.display import display, Image
```

```python
# Load the original color image
original_image = cv2.imread("/Users/hasnainkabir/Documents/Summer 2023/Digital Image Processing/LAB/Lab Practice/Lab6(Canny Edge Detector Algorithm)/Flower.jpg")
```

```python
# Step 1: Noise reduction by blurring the image using Gaussian blur
kernel_size = (5, 5)  # Should be odd and equal
standard_deviation = 0
blurred_image = cv2.GaussianBlur(original_image, kernel_size, standard_deviation)
```

```python
# Step 2: Computing the intensity gradients of the image
canny_image = cv2.Canny(blurred_image, 30, 150, apertureSize=3)
# Adjust the lower and upper threshold values (30 and 150 in this example) as needed.
```

```python
# Create a larger canvas to display both images side by side
output_image = np.hstack((original_image, cv2.cvtColor(canny_image, cv2.COLOR_GRAY2BGR)))
```

```python
# Display the concatenated image inline in the Jupyter Notebook cell
display(Image(data=cv2.imencode('.jpg', output_image)[1].tobytes()))
```
