```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('/Users/hasnainkabir/Desktop/dog.jpg', 0)  # Read as grayscale
```

```python
# Apply binary thresholding
thresh_value = 100  # Threshold value
max_value = 255  # Maximum value to be assigned

# Thresholding types
threshold_types = [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV]
thresholded_images = []
```

```python
# Apply thresholding for both types
for threshold_type in threshold_types:
    ret, thresh_image = cv2.threshold(image, thresh_value, max_value, threshold_type)
    thresholded_images.append(thresh_image)
```

```python
# Display all three images in the notebook
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

for i, thresh_img in enumerate(thresholded_images):
    plt.subplot(1, 3, i + 2)
    plt.imshow(thresh_img, cmap='gray')
    plt.title(f'Thresholded Image {i + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
